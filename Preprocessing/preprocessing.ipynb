# Converting images labels into folders for train set only

```python
import os
import shutil

# Path to the directory containing the images
image_directory = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\Train"


# Create a list of unique label names from the image filenames
label_names = set([filename.split("-")[0] for filename in os.listdir(image_directory)])

# Create a new folder for each label name
for label_name in label_names:
    folder_path = os.path.join(image_directory, label_name)
    os.makedirs(folder_path, exist_ok=True)

# Move the images to their respective label folders
for filename in os.listdir(image_directory):
    if os.path.isfile(os.path.join(image_directory, filename)):
        label_name = filename.split("-")[0]
        src_path = os.path.join(image_directory, filename)
        dst_path = os.path.join(image_directory, label_name, filename)
        shutil.move(src_path, dst_path)
```

# Converting images labels to folders for  test set only

```python
import os
import shutil

# Path to the directory containing the original images
original_folder_path = r"D:\Potent_data\Test"

# Path to the directory where you want to create the new folders
new_folder_path = r"D:\Potent_data\Newvo"

# Get a list of all the image files in the original folder
image_files = [file for file in os.listdir(original_folder_path) if file.endswith(".png")]

# Iterate through each image file
for image_file in image_files:
    # Extract the label from the image file name
    label = image_file.split("_")[5].split(".")[0]
    
    # Create the new folder path based on the label
    new_folder = os.path.join(new_folder_path, label)
    
    # Create the new folder if it doesn't already exist
    if not os.path.exists(new_folder):
        os.makedirs(new_folder)
    
    # Copy the image file to the new folder
    shutil.copy2(os.path.join(original_folder_path, image_file), new_folder)
```

# Resizing the images into 224x224 for train and test set both

```python
import os
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from PIL import Image

# Path to the directory containing the original images
original_images_directory = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\Train"

# Path to the directory where augmented images will be saved
augmented_images_directory = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\set"

# Desired image size
target_size = (224, 224)


# Maximum image size allowed
max_image_size = 178956970  # Adjust this value as needed

# Create a list of subdirectories (label folders) in the original images directory
label_folders = os.listdir(original_images_directory)

# Loop through each label folder
for label_folder in label_folders:
    # Create a directory for augmented images for the current label
    augmented_label_folder = os.path.join(augmented_images_directory, label_folder)
    os.makedirs(augmented_label_folder, exist_ok=True)

    # Get the list of original image filenames in the current label folder
    original_image_folder = os.path.join(original_images_directory, label_folder)
    original_image_filenames = os.listdir(original_image_folder)

    # Loop through each original image
    for filename in original_image_filenames:
        # Get the path of the original image
        image_path = os.path.join(original_image_folder, filename)

        # Check the image size
        image_size = os.path.getsize(image_path)
        if image_size > max_image_size:
            print(f"Skipping image {filename} due to exceeding the maximum size limit.")
            continue

        # Load the original image using OpenCV
        image = cv2.imread(image_path)

        # Resize the image to the target size
        resized_image = cv2.resize(image, target_size)

        # Save the resized image to the augmented folder
        save_path = os.path.join(augmented_label_folder, filename)
        cv2.imwrite(save_path, resized_image)

# Print a message when resizing is complete
print("Image resizing complete!")
```

# Data augmentation for train set only

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

# Path to the directory containing the original images
original_images_directory = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\set"
# Path to the directory where augmented images will be saved
augmented_images_directory =r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\augmented"

# Desired image size
target_size = (224, 224)

# Create a list of subdirectories (label folders) in the original images directory
label_folders = os.listdir(original_images_directory)

# Create an ImageDataGenerator object for augmentation
datagen = ImageDataGenerator(
    rescale=1.0/255.0,  # Rescale pixel values to [0, 1]
    rotation_range=20,  # Random rotation between -20 and 20 degrees
    width_shift_range=0.2,  # Random horizontal shift
    height_shift_range=0.2,  # Random vertical shift
    zoom_range=0.2,  # Random zoom
    horizontal_flip=True, # Random horizontal flip
    brightness_range = (0.4,0.6)
)

# Loop through each label folder
for label_folder in label_folders:
    # Create a directory for augmented images for the current label
    augmented_label_folder = os.path.join(augmented_images_directory, label_folder)
    os.makedirs(augmented_label_folder, exist_ok=True)

    # Get the list of original image filenames in the current label folder
    original_image_folder = os.path.join(original_images_directory, label_folder)
    original_image_filenames = os.listdir(original_image_folder)

    # Track the number of augmented images generated
    num_augmented_images = 0

    # Loop through each original image
    for filename in original_image_filenames:
        # Load the original image
        image_path = os.path.join(original_image_folder, filename)
        image = load_img(image_path, target_size=target_size)
        image = img_to_array(image)

        # Expand dimensions to match the expected input shape for augmentation
        image = image.reshape((1,) + image.shape)

        # Generate augmented images
        augmented_images = datagen.flow(
            x=image,
            batch_size=1,
            save_to_dir=augmented_label_folder,
            save_format="jpg"
        )

        # Generate the desired number of augmented images or until the maximum limit is reached
        while num_augmented_images < 100:
            augmented_image = augmented_images.next()[0]
            num_augmented_images += 1

            # Break the loop if the maximum limit is reached
            if num_augmented_images >= 100:
                break

    # Print the number of augmented images generated for the current label folder
    print(f"Label Folder: {label_folder} | Number of Augmented Images: {num_augmented_images}")

# Print a message when augmentation is complete
print("Data augmentation complete!")
```

```python
import os
from PIL import Image

root_dir = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\Test_resized"  # Replace with the path to the root folder containing the subfolders

# Iterate over the subfolders
for subdir in os.listdir(root_dir):
    subfolder = os.path.join(root_dir, subdir)
    if not os.path.isdir(subfolder):
        continue

    # Iterate over the images in the subfolder
    for filename in os.listdir(subfolder):
        filepath = os.path.join(subfolder, filename)

        # Open the image using PIL/Pillow
        image = Image.open(filepath)

        # Convert the image to RGBA if it's not already
        if image.mode != 'RGBA':
            image = image.convert('RGBA')

        # Create a new image with a white background
        new_image = Image.new('RGBA', image.size, (255, 255, 255))

        # Iterate over each pixel in the image
        for x in range(image.width):
            for y in range(image.height):
                # Get the pixel value
                r, g, b, a = image.getpixel((x, y))

                # Check if the pixel is black
                if r == 0 and g == 0 and b == 0:
                    # Replace the black pixel with white
                    new_image.putpixel((x, y), (255, 255, 255, a))
                else:
                    # Preserve the original pixel
                    new_image.putpixel((x, y), (r, g, b, a))

        # Save the modified image with the same filename
        new_filepath = filepath
        new_image.save(new_filepath)

        print(f"Background replaced for {filename}. Saved as {new_filepath}")
```

# Converting png images to jpg images for test set only

```python
from PIL import Image
import os

# Path to the main folder containing the subfolders with PNG images
main_folder_path = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\Test_resized"

# Path to the directory where you want to save the JPEG images
jpg_folder_path = r"D:\Clean_Reproducible_Code_For_Paper\Potent_data\jpg_test"

# Iterate through each subfolder in the main folder
for subfolder in os.listdir(main_folder_path):
    subfolder_path = os.path.join(main_folder_path, subfolder)

    # Check if the subfolder is a directory
    if os.path.isdir(subfolder_path):
        # Create a corresponding subfolder in the JPEG folder
        jpg_subfolder_path = os.path.join(jpg_folder_path, subfolder)
        if not os.path.exists(jpg_subfolder_path):
            os.makedirs(jpg_subfolder_path)

        # Iterate through each PNG image file in the subfolder
        for image_file in os.listdir(subfolder_path):
            if image_file.endswith(".png"):
                png_image_path = os.path.join(subfolder_path, image_file)
                jpg_image_path = os.path.join(jpg_subfolder_path, image_file.replace(".png", ".jpg"))

                # Open the PNG image using PIL/Pillow
                image = Image.open(png_image_path)

                # Create a new image with a white background
                new_image = Image.new("RGB", image.size, (255, 255, 255))

                # Composite the PNG image on the new white background
                new_image.paste(image, (0, 0), mask=image.convert("RGBA").split()[3])

                # Convert and save the image as JPEG
                new_image.convert("RGB").save(jpg_image_path, "JPEG")
```
